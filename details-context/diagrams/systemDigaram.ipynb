```python
# systemDiagram.py

from hmac import digest
from diagrams import Cluster, Diagram, Edge
from diagrams.aws.compute import EC2
from diagrams.aws.database import Database
from diagrams.aws.network import ELB
from diagrams.custom import Custom
from diagrams.aws.general import Users
# from diagrams.gcp.database import Database
from diagrams.aws.compute import Compute
from diagrams.gcp.network import Router

# credits
'''
<a href="https://www.flaticon.com/free-icons/browser" title="browser icons">Browser icons created by Icon home - Flaticon</a>
<a href="https://www.flaticon.com/free-icons/interface" title="interface icons">Interface icons created by juicy_fish - Flaticon</a>
<a href="https://www.flaticon.com/free-icons/input" title="input icons">Input icons created by Royyan Wijaya - Flaticon</a>
<a href="https://www.flaticon.com/free-icons/sensor" title="sensor icons">Sensor icons created by Prosymbols Premium - Flaticon</a>
<a href="https://www.flaticon.com/free-icons/ui" title="ui icons">Ui icons created by muhammad atho' - Flaticon</a>
<a href="https://www.flaticon.com/free-icons/display" title="display icons">Display icons created by Ralf Schmitzer - Flaticon</a>
<a href="https://www.flaticon.com/free-icons/code" title="code icons">Code icons created by Icongeek26 - Flaticon</a>
'''

graph_attr = {
    "label": "System",
    "fontsize": "40",
    "bgcolor": "white",
    "pad": "0.1"
}

with Diagram(show=False, filename="systemDiagram-1.0", graph_attr=graph_attr) as system:
    # create the nodes
    user = Users("user")

    with Cluster("Front-end"):
        display = Custom("Output Playlist", "./logos/display.png")
        input = Custom("Input Songs", "./logos/inputText.png")

    website = Custom("main program", "./logos/web-programming.png")

    with Cluster("Backend"):
        # with Cluster("Boiling the Frog"):
        sortSongs = Custom("Sort Songs", "./logos/microchipSmall.png")
        addSongs = Custom("Add Songs", "./logos/microchipPlus.png")
        interfaceSpotify = Custom("Interface With Spotify", "./logos/software-application.png")
    
    spotify = Custom("Spotify", "./logos/spotify-2-logo-png-transparent.png")

    # main = 

    # build the diagram
    # frontend
    user >> input >> website
    user << display << website
    

    # backend
    website << sortSongs 
    website << addSongs
    
    addSongs << interfaceSpotify
    sortSongs << interfaceSpotify
    
    #sortSongs >> addSongs
    interfaceSpotify << spotify

system



```

```python
# credits
'''
<a href = "https://www.flaticon.com/free-icons/cluster-computing" title = "cluster computing icons" > Cluster computing icons created by IconBaandar - Flaticon < /a >
'''

graph_attr_subsystems = {
    "label": "Subsytems",
    "fontsize": "40",
    "bgcolor": "white",
    "pad": "0.5"
}


with Diagram(show=False, filename="systemDiagramSubsystems-1.0", graph_attr=graph_attr_subsystems) as subSystem:
    # create the nodes

    with Cluster("Backend"):
        plotSongs = Custom("Plot Songs", "./logos/cluster.png")
        sortSongs = Custom("Songs Sorted into Playlist", "./logos/microchipSmall.png")
        addSongs = Custom("Smooth Playlist of Songs", "./logos/microchipPlus.png")
        #genreClusters = Custom("genre clusters", "./logos/cluster.png")


    website = Custom("main program", "./logos/web-programming.png")    
    interfaceSpotify = Custom(
            "Interface With Spotify", "./logos/software-application.png")

    # create edges
    plotSongs <<  Edge(label="Plot song list in genra clusters.") << website
    plotSongs >> Edge(label="Plot a path through clusters including every song from list." ) >> sortSongs
    sortSongs >> Edge(label="If path goes through cluster without a song, add song from that cluster.", style="dashed") >> addSongs
    website << Edge(label="Send Playlist") << addSongs
    
    plotSongs << Edge(label="use data imported from spotify") << interfaceSpotify


subSystem
```
